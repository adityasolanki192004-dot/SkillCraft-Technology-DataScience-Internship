```python
import pandas as pd

# Create sample traffic accident data
data = {
    "Weather": ["Clear", "Rain", "Fog", "Clear", "Rain", "Clear", "Fog", "Rain"],
    "Road_Condition": ["Dry", "Wet", "Wet", "Dry", "Wet", "Dry", "Wet", "Wet"],
    "Time": ["Morning", "Night", "Evening", "Afternoon", "Night", "Morning", "Evening", "Night"]
}

df = pd.DataFrame(data)

print(df)
```

```python
import matplotlib.pyplot as plt

df["Weather"].value_counts().plot(kind="bar")

plt.title("Accidents by Weather")
plt.xlabel("Weather")
plt.ylabel("Number of Accidents")

plt.show()
```

```python
df["Road_Condition"].value_counts().plot(kind="bar")

plt.title("Road Conditions")
plt.xlabel("Road Condition")
plt.ylabel("Count")

plt.show()
```

```python
df["Time"].value_counts().plot(kind="bar")

plt.title("Accidents by Time")
plt.xlabel("Time")
plt.ylabel("Count")

plt.show()
```
